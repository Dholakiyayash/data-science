```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this 
# generated data plot a scatter plot.


# A scatter plot is a type of data visualization that displays individual data points as dots on
# a two-dimensional Cartesian coordinate system. 

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.title('scatter plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# Use the following data:


# The subplot() function in Matplotlib is used to create multiple plots within the same figure. 
# It allows you to arrange multiple subplots in a grid format, which is helpful when you want to 
# compare and visualize different aspects of the data side by side. 

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title('plot 1')

plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title('plot 2')

plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title('plot 3')

plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.title('plot 4')


```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and 
# a horizontal bar plot.


# A bar plot is a type of data visualization that represents categorical data with rectangular bars. 
# Each bar's length or height corresponds to the value of the category it represents. Bar plots are 
# commonly used to compare and display the relative frequencies, counts, or numerical values of 
# different categories in a dataset.

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.


# A box plot, also known as a box-and-whisker plot, is a data visualization technique that provides a
# concise summary of the distribution of a dataset. It displays key statistical measures and
# highlights potential outliers. The box plot is particularly useful for comparing multiple datasets 
# or identifying the spread and skewness of a single dataset.


# Box plots offer a concise summary of important statistical measures such as the median, quartiles
# and range. This summary is particularly useful when presenting data to an audience or including it 
# in reports.

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data=[box1,box2]
data = [box1, box2]
plt.boxplot(data)
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot Example')
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.show()
```
