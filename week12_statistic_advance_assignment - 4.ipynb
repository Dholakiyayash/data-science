**Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.**

Difference between t-test and z-test:

Sample Size: The primary difference between a t-test and a z-test is the sample size. A t-test is used when the sample size is small (typically less than 30) and the population standard deviation is unknown. A z-test is used when the sample size is larger and the population standard deviation is known or can be reasonably estimated.

Standard Deviation: In a t-test, the sample standard deviation is used to estimate the population standard deviation. In a z-test, the population standard deviation is either known or assumed to be known.

Distribution: A t-test uses the t-distribution, which has heavier tails compared to the standard normal distribution (z-distribution). The t-distribution becomes closer to the z-distribution as the sample size increases.

Example scenarios:

T-Test Example:
Let's say a pharmaceutical company wants to test if a new drug reduces blood pressure. They randomly select 20 patients and measure their blood pressure before and after taking the drug. Since the sample size is small and the population standard deviation is unknown, a paired t-test would be appropriate to determine if there's a significant difference in blood pressure before and after taking the drug.

Z-Test Example:
Suppose an online retailer wants to compare the average purchase amount of customers who visit their website on weekends and weekdays. They collect purchase data from a large sample of customers. Since the sample size is large and the retailer has a good estimate of the population standard deviation, they can use a z-test to determine if there's a significant difference in the average purchase amount between the two groups.

**Q2: Differentiate between one-tailed and two-tailed tests.**

One-Tailed Test:
In a one-tailed test, the alternative hypothesis is focused on only one direction of the distribution. This means that you are interested in detecting an effect in one specific direction (either greater than or less than). The critical region is located in one tail of the distribution.

There are two types of one-tailed tests:

Left-Tailed Test: The critical region is in the left tail of the distribution. This is used when you want to determine if the sample statistic is significantly less than a certain value.

Right-Tailed Test: The critical region is in the right tail of the distribution. This is used when you want to determine if the sample statistic is significantly greater than a certain value.

Two-Tailed Test:
In a two-tailed test, the alternative hypothesis is looking for effects in both directions of the distribution. This means that you are interested in detecting any significant difference, whether it's greater than or less than. The critical region is divided between the two tails of the distribution.


**Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.**

Type 1 Error (False Positive):
A Type 1 error occurs when you reject a true null hypothesis. In other words, you conclude that there is an effect or difference when, in reality, there is none. This is also known as a "false positive" or an "alpha error." It's the mistake of seeing something that isn't actually there.

Example Scenario for Type 1 Error:
Imagine a medical test for a rare disease. The null hypothesis (H0) is that the person is not infected with the disease. The alternative hypothesis (Ha) is that the person is infected. If the test incorrectly indicates that a healthy person has the disease (rejects H0 when it's true), this is a Type 1 error. The person is unnecessarily treated for a disease they don't have.

Type 2 Error (False Negative):
A Type 2 error occurs when you fail to reject a false null hypothesis. In other words, you conclude that there is no effect or difference when, in reality, there is one. This is also known as a "false negative" or a "beta error." It's the mistake of not seeing something that is actually there.

Example Scenario for Type 2 Error:
Continuing with the medical test example, if the test fails to detect the disease in an infected person (fails to reject H0 when it's false), this is a Type 2 error. The person doesn't receive appropriate treatment because the test didn't catch the disease.

In hypothesis testing, the level of significance (α) is set to control the probability of making a Type 1 error. When you set a lower α, you reduce the chance of a Type 1 error, but you increase the chance of a Type 2 error.

**Q4: Explain Bayes's theorem with an example.**

Bayes's Theorem is a fundamental concept in probability theory that allows us to update our beliefs about the probability of an event based on new evidence. It provides a way to calculate the conditional probability of an event A given that event B has occurred, taking into account the prior probability of event A and the likelihood of event B given event A.

P(A/ B)= P(A)*P(B/A) / P(B)

A, B	=	events

P(A|B)	=	probability of A given B is true

P(B|A)	=	probability of B given A is true

P(A), P(B)	=	the independent probabilities of A and B


Example:
Suppose you have two dice, Die A and Die B. You know that Die A is fair (each side has an equal chance of landing face up), but you're not sure about Die B. You're told that Die B is biased and has a higher chance of landing on an even number.

You want to calculate the probability that you rolled Die A given that you rolled an even number (2, 4, or 6).

Given:

Die A is fair.

Die B is biased with a higher chance of landing on an even number.

Prior Probability of rolling Die A (P(A)): 0.5

Prior Probability of rolling Die B (P(¬A)): 0.5

Probability of rolling an even number on Die A (P(B|A)): 0.5

Probability of rolling an even number on Die B (P(B|¬A)): 0.8

Probability of rolling an even number (P(B)):

Since P(B) is the probability of rolling an even number, it can be calculated using the probabilities of rolling an even number on Die A and Die B:

P(B)=P(B∣A)⋅P(A)+P(B∣¬A)⋅P(¬A)

P(B)=0.5*0.5+0.8*0.5

P(B)=0.65

Finally, substituting all values into Bayes's Theorem:

P(A∣B)= P(B∣A)*P(A) / P(B)

P(A∣B)= 0.5*0.5 / 0.65

P(A∣B)≈0.3846

So, even though Die B has a higher chance of landing on an even number, the probability that you rolled Die A given that you rolled an even number is approximately 0.3846 (or about 38.46%). This example demonstrates how Bayes's Theorem allows us to update our beliefs based on new evidence.

Calculating a confidence interval involves estimating a range of values around a sample statistic to capture the likely range of the true population parameter. Here's a concise explanation with an example:

Step-by-Step: Calculating a Confidence Interval

Collect Sample Data: Obtain a representative sample from the population.

Calculate Sample Statistic: Calculate the sample mean (for estimating population mean) or sample proportion (for estimating population proportion).

Determine Confidence Level: Choose a confidence level, such as 95%, 99%, etc. This represents the level of certainty you want.

Identify Critical Value: Determine the critical value associated with the chosen confidence level. It's based on the distribution (e.g., z-distribution for large samples, t-distribution for small samples).

Calculate Standard Error: Compute the standard error of the sample statistic. It accounts for the variability in the sample.

Calculate Confidence Interval: Use the formula: Confidence Interval = Sample Statistic ± (Critical Value) × (Standard Error).

Example:
Suppose you want to estimate the average score of a population in a test. You collect a sample of 50 scores and find the sample mean to be 85. The population standard deviation is known to be 10. You want to calculate a 95% confidence interval for the true population mean score.

Given:

Sample Mean (x̄) = 85

Sample Size (n) = 50

Population Standard Deviation (σ) = 10

Confidence Level = 95%

Z-Critical Value for 95% confidence level ≈ 1.96 (from standard normal distribution)

Standard Error (SE) = σ/sqrt(n) =10/sqrt(50) ≈1.414

Confidence Interval:

Confidence Interval=Sample Mean±(Critical Value)×(Standard Error)

Confidence Interval=85±(1.96)×(1.414)

Confidence Interval=(82.23,87.77)

**Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error?Provide an example of a scenario where a larger sample size would result in a smaller margin of error.**

The margin of error in a confidence interval is the range around the point estimate (like the sample mean) that quantifies the uncertainty associated with estimating a population parameter. It reflects how much the point estimate might vary due to sampling variability. The margin of error indicates the precision of the estimate; a smaller margin of error indicates a more precise estimate.

Mathematically, the margin of error is calculated as:

Margin of Error=Critical Value×Standard Error

Where the critical value is determined by the desired confidence level and the distribution (z-score for normal distribution, t-score for t-distribution), and the standard error accounts for the variability due to the sample size and standard deviation.

Sample Size and Margin of Error:
Sample size has an inverse relationship with the margin of error. As the sample size increases, the margin of error decreases, leading to a more precise estimate. This is because larger samples tend to provide more representative and reliable information about the population.

Example Scenario:
Suppose a polling agency wants to estimate the proportion of voters in a city who support a particular candidate. They conduct two separate polls, each with a different sample size.

Poll 1: Sample size n=200,
 Proportion of supporters p=0.60

Poll 2: Sample size n=1000,
 Proportion of supporters p=0.60

For both polls, the point estimate of the proportion of supporters is 0.60. However, the margin of error for Poll 2 will be smaller due to its larger sample size. This means that the confidence interval for Poll 2 will be narrower, and the estimate is more precise.

For example, let's say both polls are conducted with a 95% confidence level. The critical value for a 95% confidence level from the z-table is approximately 1.96.

Margin of Error Calculation:

For Poll 1:

Margin of Error=1.96 × sqrt(0.60×(1−0.60)/200)
 ≈0.0675

For Poll 2:

Margin of Error=1.96 × sqrt(0.60×(1−0.60)/1000)
 ≈0.0309

In this scenario, Poll 2 has a smaller margin of error (0.0309) compared to Poll 1 (0.0675), indicating a more precise estimate of the proportion of supporters.

So, a larger sample size generally leads to a smaller margin of error and a more accurate estimation of the population parameter.

The z-score (also known as the standard score) measures how many standard deviations a data point is away from the mean of the population. It's a way to standardize data and make comparisons across different distributions. The formula to calculate the z-score is:


Data Point = 75
Population Mean = 70
Population Standard Deviation = 5

z= Data Point−Population Mean/Population Standard Deviation

=75-70 / 5=1

Interpretation:

A z-score of 1 means that the data point (75) is 1 standard deviation above the population mean (70). In other words, it's relatively higher than the average value in the population by one standard deviation. This indicates that the data point is in the upper part of the distribution, away from the center. The positive sign of the z-score indicates that the data point is above the mean.

Z-scores are particularly useful for comparing data points from different distributions or for identifying outliers. A z-score of 1 is considered a moderate deviation from the mean, and it's often used to determine how unusual or significant a particular data point is within the context of the population distribution.
