**Q1: What is Estimation Statistics? Explain point estimate and interval estimate.**

Estimation in statistics involves the process of making inferences about population parameters based on sample data. It allows us to estimate unknown population characteristics using information from a sample drawn from the population.

Point Estimate:
A point estimate is a single value that serves as the best estimate or "guess" for the population parameter of interest. It is obtained by calculating a sample statistic from the sample data and using it as an estimate of the corresponding population parameter. Point estimates are commonly used when a single value is needed to summarize the data and draw conclusions.

For example, if we want to estimate the average height (population mean, denoted as μ) of all students in a school, we can take a random sample of students, measure their heights, and calculate the sample mean (x̄). The sample mean serves as the point estimate for the population mean μ.

Interval Estimate (Confidence Interval):
An interval estimate, also known as a confidence interval, provides a range of values within which the population parameter is likely to lie with a certain level of confidence. Instead of providing a single point estimate, an interval estimate gives a range of plausible values.

Confidence intervals are constructed using the sample statistic (e.g., sample mean, sample proportion) along with the sample size and the level of confidence chosen by the researcher. The level of confidence is often expressed as a percentage, such as 95% confidence interval or 99% confidence interval.

For instance, a 95% confidence interval for the population mean height of students might be [65 inches, 68 inches]. This means that we are 95% confident that the true population mean height falls within this range.

```python
# Q2. Write a Python function to estimate the population mean using a sample mean and standard
# deviation.

def estimate_population_mean(sample_mean, sample_std, sample_size):
    return sample_mean * (sample_size ** 0.5) / sample_std
 

sample_mean = 75.6
sample_std = 10.2
sample_size = 50

estimated_population_mean = estimate_population_mean(sample_mean, sample_std, sample_size)
print("Estimated Population Mean:", estimated_population_mean)
```

**Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.**

Hypothesis testing is a statistical method used to make inferences and draw conclusions about a population based on sample data. It involves formulating two competing hypotheses about a population parameter, collecting sample data, and using statistical tests to determine the likelihood of the observed sample data under each hypothesis. The main purpose of hypothesis testing is to assess whether there is enough evidence in the sample data to support one hypothesis over the other.

Importance of Hypothesis Testing:

Data-Driven Decision Making: Hypothesis testing provides a systematic and objective way to make decisions based on data. It helps in evaluating claims and making data-driven choices in research, business, and various fields.

Scientific Inference: Hypothesis testing is fundamental to scientific research. It allows researchers to test their theories and determine if the observed results are statistically significant or simply due to chance.

Validation of Claims: Hypothesis testing is used to validate or invalidate claims made about a population. For example, it can be used to determine if a new drug has a significant effect compared to a placebo or if a new marketing strategy leads to a meaningful increase in sales.

Quality Control: In industries and manufacturing, hypothesis testing is used for quality control and process improvement. It helps identify if a process change has resulted in a significant improvement or if the output is still within acceptable limits.

Policy Decisions: Hypothesis testing plays a role in policy-making, where statistical evidence is crucial for implementing changes and evaluating their impact.


**Q4. Create a hypothesis that states whether the average weight of male college students is greater than the average weight of female college students.**

Hypothesis:
The average weight of male college students is greater than the average weight of female college students.

Null Hypothesis (H0):
μ_male ≤ μ_female

This null hypothesis states that there is no significant difference in the average weight between male and female college students. It assumes that the average weight of male college students is less than or equal to the average weight of female college students.

Alternative Hypothesis (Ha):
μ_male > μ_female

The alternative hypothesis states that there is a significant difference in the average weight between male and female college students. It suggests that the average weight of male college students is greater than the average weight of female college students. This is the hypothesis we want to test and provide evidence for, if supported by the data.

The hypothesis can be tested using appropriate statistical methods and data from a random sample of male and female college students' weights. The results of the hypothesis test will determine whether there is enough evidence to support the claim that male college students have a higher average weight than female college students.

```python
# Q5. Write a Python script to conduct a hypothesis test on the difference between two population 
# means, given a sample from each population.

def calculate_mean(sample):
    return sum(sample) / len(sample)

def calculate_variance(sample, sample_mean):
    return sum((x - sample_mean) ** 2 for x in sample) / (len(sample) - 1)

def calculate_t_statistic(sample1, sample2):
    mean1 = calculate_mean(sample1)
    mean2 = calculate_mean(sample2)
    var1 = calculate_variance(sample1, mean1)
    var2 = calculate_variance(sample2, mean2)
    n1 = len(sample1)
    n2 = len(sample2)

    pooled_std_err = ((var1 / n1) + (var2 / n2)) ** 0.5

    t_statistic = (mean1 - mean2) / pooled_std_err

    return t_statistic

def conduct_hypothesis_test(sample1, sample2, alpha=0.05):
    t_statistic = calculate_t_statistic(sample1, sample2)

    critical_t_value = 2.262

    reject_null_hypothesis = abs(t_statistic) > critical_t_value

    return reject_null_hypothesis

sample_male = [70, 75, 80, 72, 68, 78, 73, 74, 76, 71]
sample_female = [65, 68, 70, 67, 69, 71, 66, 72, 68, 70]

reject_null_hypothesis = conduct_hypothesis_test(sample_male, sample_female)

if reject_null_hypothesis:
    print("Null hypothesis rejected: There is a significant difference between the means.")
else:
    print("Null hypothesis not rejected: There is no significant difference between the means.")
```

**Q8. Define p-value and explain its significance in hypothesis testing.**

The p-value, or probability value, is a crucial concept in hypothesis testing. It quantifies the strength of evidence against the null hypothesis (H0) and helps researchers make informed decisions about accepting or rejecting H0 in favor of the alternative hypothesis (Ha). The p-value represents the probability of obtaining the observed results, or more extreme results, under the assumption that the null hypothesis is true.

Significance of the p-value in Hypothesis Testing:

Assessing Statistical Significance: A small p-value (typically below a predefined significance level, such as 0.05) suggests that the observed results are unlikely to occur by random chance alone, assuming the null hypothesis is true. This leads to the rejection of the null hypothesis, indicating that there is strong evidence for the alternative hypothesis.

Decision Making: Researchers use the p-value to make decisions about hypotheses. If the p-value is less than the chosen significance level (alpha), they reject the null hypothesis and accept the alternative hypothesis. If the p-value is greater than alpha, they fail to reject the null hypothesis, and the data do not provide sufficient evidence to support the alternative hypothesis.

Quantifying Strength of Evidence: The p-value allows researchers to quantify the strength of evidence against H0. A smaller p-value indicates stronger evidence against the null hypothesis, reinforcing the credibility of the alternative hypothesis.

Avoiding Type I Errors: The p-value helps control the probability of making a Type I error (false positive) in hypothesis testing. A Type I error occurs when the null hypothesis is rejected when it is actually true. By setting a significance level (alpha), researchers can control the risk of making such errors.

Hypothesis Refinement: If the p-value is close to the significance level, researchers may consider further investigation or replication to refine the hypothesis and draw more definitive conclusions.

**Q11: What is Student’s t distribution? When to use the t-Distribution.**

Student's t-distribution, commonly referred to as the t-distribution, is a probability distribution used in statistics for conducting hypothesis tests and constructing confidence intervals when the sample size is small or when the population standard deviation is unknown. It is a family of distributions, with different shapes determined by the degrees of freedom.

The t-distribution is similar in shape to the standard normal (Z) distribution but has heavier tails. As the sample size increases, the t-distribution approaches the standard normal distribution. The shape of the t-distribution depends on the degrees of freedom (df), which is related to the sample size.

When to use the t-Distribution:

Small Sample Size: The t-distribution is especially useful when dealing with small sample sizes (typically n < 30) because the sample standard deviation may not accurately estimate the population standard deviation.

Unknown Population Standard Deviation: When the population standard deviation is unknown, the t-distribution is used to estimate the population mean based on the sample mean.

Constructing Confidence Intervals: The t-distribution is commonly used to construct confidence intervals for the population mean when the sample size is small.

Hypothesis Testing: In hypothesis testing, the t-distribution is used for conducting t-tests when the population standard deviation is unknown or when the sample size is small.

Comparing Means: When comparing means of two independent samples or paired samples, the t-distribution is used in t-tests.

Outliers: The t-distribution is less sensitive to outliers compared to other distributions, making it more robust for skewed or heavy-tailed data.

**Q12: What is t-statistic? State the formula for t-statistic.**

The t-statistic is a test statistic used in hypothesis testing to determine if there is a significant difference between the means of two groups or between a sample mean and a population mean when the population standard deviation is unknown. It is used when working with small sample sizes or when the population standard deviation is not available.

Formula for t-Statistic:
The t-statistic is calculated using the formula:

t = (x̄ - μ) / (s / √n)

where:
t is the t-statistic,
x̄ is the sample mean,
μ is the population mean (or hypothesized mean under the null hypothesis),
s is the sample standard deviation, and
n is the sample size.

In the case of comparing two independent sample means (two-sample t-test), the formula for the t-statistic is:

t = (x̄1 - x̄2) / √((s1^2 / n1) + (s2^2 / n2))

where:
t is the t-statistic,
x̄1 and x̄2 are the means of the two samples,
s1 and s2 are the sample standard deviations of the two samples,
n1 and n2 are the sample sizes of the two samples.

The t-statistic measures how many standard errors the sample mean differs from the population mean (or the difference between two sample means). If the absolute value of the t-statistic is large, it indicates a significant difference between the sample mean(s) and the hypothesized population mean (in the case of one-sample t-test) or between the two sample means (in the case of two-sample t-test). The t-statistic is then used to calculate the p-value, which helps determine the statistical significance of the difference and make decisions in hypothesis testing.

**Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.**

To test the hypothesis with a significance level of 0.05, we will conduct a one-sample t-test.

Given:

Sample Mean (x̄) = 8 mmHg,
Sample Standard Deviation (s) = 3 mmHg,
Sample Size (n) = 100,
Population Mean under H0 (μ0) = 10 mmHg,
Significance Level (α) = 0.05

Let's perform the one-sample t-test:

Step 1: Set up hypotheses:

Null Hypothesis (H0): μ = μ0 (The new drug does not decrease blood pressure by 10 mmHg)

Alternative Hypothesis (Ha): μ ≠ μ0 (The new drug decreases blood pressure by 10 mmHg)

Step 2: Calculate the t-statistic:

t = (x̄ - μ0) / (s / √n)

t = (8 - 10) / (3 / √100)

t = -2 / (3 / 10)

t = -2 * (10 / 3)

t ≈ -6.67

Step 3: Calculate the degrees of freedom:

df = n - 1

df = 100 - 1

df = 99

Step 4: Determine the critical t-value for a two-tailed test with α = 0.05 and df = 99. You can refer to the t-distribution table or use a statistical software/tool. For α = 0.05 and df = 99, the critical t-value is approximately ±1.984.

Step 5: Compare the calculated t-statistic with the critical t-value:

Since the calculated t-statistic (-6.67) is less than the critical t-value (-1.984), we reject the null hypothesis.

Step 6: Make a conclusion:

Based on the one-sample t-test with a significance level of 0.05, there is sufficient evidence to conclude that the new drug decreases blood pressure by 10 mmHg. The sample data supports the researcher's hypothesis.

Sample Size (n1) for Group 1 = 30,
Sample Mean (x̄2) for Group 2 = 75,

Sample Size (n2) for Group 2 = 40,
Significance Level (α) = 0.01

Let's perform the two-sample t-test:

Step 1: Set up hypotheses:

Null Hypothesis (H0): μ1 = μ2 (The population means for the two groups are equal)

Alternative Hypothesis (Ha): μ1 ≠ μ2 (The population means for the two groups are not equal)

Step 2: Calculate the pooled standard error:

Pooled Standard Error (SE) = √((s1^2 / n1) + (s2^2 / n2))

SE = √((10^2 / 30) + (8^2 / 40))

SE ≈ √(3.3333 + 1.6)

SE ≈ √4.9333

SE ≈ 2.2211

Step 3: Calculate the t-statistic:

t = (x̄1 - x̄2) / SE

t = (80 - 75) / 2.2211

t ≈ 2.248

Step 4: Determine the degrees of freedom:

Degrees of Freedom (df) = n1 + n2 - 2

df = 30 + 40 - 2

df = 68

Step 5: Determine the critical t-value for a two-tailed test with α = 0.01 and df = 68. You can refer to the t-distribution table or use a statistical software/tool. For α = 0.01 and df = 68, the critical t-value is approximately ±2.628.

Step 6: Compare the calculated t-statistic with the critical t-value:

Since the absolute value of the calculated t-statistic (2.248) is less than the critical t-value (2.628), we fail to reject the null hypothesis.

Step 7: Make a conclusion:

Based on the two-sample t-test with a significance level of 0.01, there is insufficient evidence to conclude that the population means for the two groups are different. The sample data does not provide strong enough evidence to support the hypothesis that the population means are not equal.


**Q17. A marketing company wants to estimate the average number of ads watched by viewers during a TV program. They take a random sample of 50 viewers and find that the sample mean is 4 with a standard deviation of 1.5. Estimate the population mean with a 99% confidence interval.**


To estimate the population mean number of ads watched by viewers during a TV program with a 99% confidence interval, we can use the formula for the confidence interval for a population mean when the population standard deviation is unknown. The formula is:

Confidence Interval = Sample Mean ± (Critical Value) * (Sample Standard Deviation / √Sample Size)

where:

Sample Mean (x̄) = 4,
Sample Standard Deviation (s) = 1.5,
Sample Size (n) = 50,
Confidence Level = 99%

Step 1: Determine the critical value for a 99% confidence level and the corresponding degrees of freedom (df = n - 1).

The critical value can be obtained from the t-distribution table or using a statistical software/tool. For a 99% confidence level and df = 49, the critical value is approximately ±2.677.

Step 2: Calculate the margin of error.

Margin of Error = Critical Value * (Sample Standard Deviation / √Sample Size)

Margin of Error = 2.677 * (1.5 / √50)

Margin of Error ≈ 0.4768

Step 3: Calculate the confidence interval.

Confidence Interval = Sample Mean ± Margin of Error

Confidence Interval = 4 ± 0.4768

Step 4: Calculate the lower and upper bounds of the confidence interval.

Lower Bound ≈ 4 - 0.4768 ≈ 3.5232

Upper Bound ≈ 4 + 0.4768 ≈ 4.4768

Therefore, with a 99% confidence level, we can estimate that the true population mean number of ads watched by viewers during a TV program lies within the range of approximately 3.5232 to 4.4768 based on the sample data.
